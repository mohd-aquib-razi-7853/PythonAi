```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Ploting the data

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Angel")
plt.ylabel("Sine value")
plt.title("Sine Wave")
plt.show()
```

```python
#parabola 
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Ploting the data

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Angel")
plt.ylabel("Sine value")
plt.title("Sine Wave")
plt.show()
```

```python
#parabola 
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,cmap='Blues',c=z)
plt.show()
```

```python
#house price dataset
from sklearn.datasets import fetch_openml
```

```python
housing = fetch_openml(name='boston', version=1)
print(housing)
```

```python
houseDf = pd.DataFrame(housing.data, columns=housing.feature_names)
houseDf["PRICE"] = housing.target
houseDf.head()
```

```python
sns.displot(houseDf['PRICE'])
```

```python
sns.distplot(houseDf['PRICE'])
```

Co relation 
1. Positive
2. Negative

```python
corr = houseDf.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr, cmap="Blues",fmt='.1f',annot=True , square=True,cbar=True)
```

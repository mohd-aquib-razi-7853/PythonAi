# Vector operation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

### Ploting the vector

```python
plt.quiver(0,0,4,6)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1)
plt.quiver(0,0,-5,-2,scale_units='xy', angles='xy',scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Addition of two vector

```python
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])
s = vec1+vec2
print(s)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1)
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy',scale=1, color='r')
plt.quiver(0,0,5,1,scale_units='xy', angles='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Subtraction of 2 vector

```python
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])
d = vec1-vec2
print(d)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1)
plt.quiver(0,0,3,-2,scale_units='xy', angles='xy',scale=1, color='r')
plt.quiver(0,0,-1,5,scale_units='xy', angles='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

### Multiply a scalar with vector

```python
vect1 = np.asarray([0,0,2,3])

vect2 = -0.5*vect1
print(vect2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy', angles='xy',scale=1)
plt.quiver(0,0,-1,-1.5,scale_units='xy', angles='xy',scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

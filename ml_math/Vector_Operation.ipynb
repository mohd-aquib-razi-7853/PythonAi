```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

Ploting the vector

```python
plt.quiver(0,0,4,6)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy', angles='xy',scale=1)
plt.quiver(0,0,-5,-2,scale_units='xy', angles='xy',scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vector

```python
vec1 = np.asarray([0,0,2,3])
vec2 = np.asarray([0,0,3,-2])
s = vec1+vec2
print(s)
```

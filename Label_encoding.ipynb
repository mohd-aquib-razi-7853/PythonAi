Labeling ENcoding

```python
#importing the library
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_df = pd.read_csv("breast_cancer_data.csv")
```

```python
#print first 5 rows 
cancer_df.head()
```

```python
cancer_df.value_counts('diagnosis')
```

```python
#loading the label encoder
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_df.diagnosis)
```

```python
#appending the labels 
cancer_df["target"] = labels
```
